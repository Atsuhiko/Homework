「join_data_1」に「customer_master」と「item_master」をそれぞれ"customer_id"と"item_id"をキーにして結合させ「join_data2.csv」を保存してください。
結合前と結合後のデータフレイムの構造（項目数、行数、データ型、欠損値の有無など）を調べてください。
（ヒント）Pandas 「merge」メソッドを使う。

```python
import pandas as pd
```

```python
customer_master = pd.read_csv('customer_master.csv')
customer_master.head()
```

```python
item_master = pd.read_csv('item_master.csv')
item_master.head()
```

```python
join_data_1 = pd.read_csv('join_data_1.csv')
join_data_1.head()
```

```python
join_data_2 = pd.merge(join_data_1, customer_master, on="customer_id", how="left")
join_data_2 = pd.merge(join_data_2, item_master, on="item_id", how="left")
join_data_2.head()
```

```python
print(len(customer_master))
print(len(item_master))
print(len(join_data_2))
```

```python
join_data_2.info()
```

```python
# 欠損値のカウント(2)
join_data_2.isnull().sum()
```

```python
join_data_2.to_csv("join_data_2.csv",index=False)
```
